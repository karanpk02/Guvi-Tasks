```python
# importing pandas
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
csv1 = 'https://raw.githubusercontent.com/karanpk02/Guvi-Tasks/main/college_1.csv'
csv2 = 'https://raw.githubusercontent.com/karanpk02/Guvi-Tasks/main/college_2.csv'
```

```python
df1 = pd.read_csv(csv1)
df1.head()
```

```python
df2 = pd.read_csv(csv2)
df2.head()
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
df = pd.concat(map(pd.read_csv, [csv1,csv2]), ignore_index=True)
df
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv

```python
filtered_df1 = df[df['CodeKata Score'] >= 15000]
filtered_df1.to_csv('Exceeded expectations.csv')
filtered_df1
```

```python
filtered_df3 = df[(7000<= df['CodeKata Score']) & ( df['CodeKata Score'] < 10000)]
filtered_df3.to_csv('Needs_Improvement.csv')
filtered_df3
```

```python
filtered_df4 = df[df['CodeKata Score'] < 7000]
filtered_df4.to_csv('Unsatisfactory.csv')
filtered_df4
```

###### Department wise codekata performence (pie chart)

```python
 telecom6 = df.groupby('Department')['CodeKata Score'].mean() #orsum
#  telecom6 = telecom6.div(telecom6.sum().astype(float)).round(2)*100
 telecom6 
```

```python
plt.figure(figsize=(7, 7), dpi=150)
telecom6.plot.pie(colormap='brg', explode=(0, .1, 0), autopct='%0.0f%%', shadow=True,  fontsize=14) 
plt.title('Department wise codekata performence')
plt.legend()
# Show plot,  title = "Fig.No. 7 : PIE CHART WITH PERCENTAGE", 
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)
